```python
import csv
import matplotlib.pyplot as plt

data = []
data_dict = {}
list1 = []
list2 = []
list3 = []
list4 = []
list5 = []
list6 = []
list7 = []
list8 = []
list9 = []
list10 = []
list11 = []
list12 = []
year = []
months = []
with open('temp.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=',')
    line = 0
    for row in csv_reader:
        if line == 0:
            line = line +1
            months.append(row[1:])
        else:
            data.append([float(i) for i in row[1:]])
            year.append(int(row[0]))
            
#place data for different months on rows and year in columns
for row in data:
    list1.append(row[0])
    list2.append(row[1])
    list3.append(row[2])
    list4.append(row[3])
    list5.append(row[4])
    list6.append(row[5])
    list7.append(row[6])
    list8.append(row[7])
    list9.append(row[8])
    list10.append(row[9])
    list11.append(row[10])
    list12.append(row[11])
       

data_dict['Jan'] = list1
data_dict['Feb'] = list2
data_dict['Mar'] = list3
data_dict['Apr'] = list4
data_dict['May'] = list5
data_dict['Jun'] = list6
data_dict['Jul'] = list7
data_dict['Aug'] = list8
data_dict['Sep'] = list9
data_dict['Oct'] = list10
data_dict['Nov'] = list11
data_dict['Dec'] = list12

```

```python
#Average temperature from 1901 to 2017
import statistics
color = ['aqua','blue','brown','black','red','green','cyan','magenta','coral','darkgreen','gold','yellow']
plt.figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
for (month,c) in zip(data_dict.keys(),color):
    plt.plot(year,data_dict[month],color=c,label=month)

plt.xlabel('Year')      
plt.ylabel('Average Temperature')      
plt.legend(loc='upper center',bbox_to_anchor=(0.5, 1.05),ncol=6, fancybox=True, shadow=True)
```
